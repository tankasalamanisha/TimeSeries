## Importing dependencies

```python
import pandas as pd
```

## Reading data

```python
df = pd.read_csv('SBUX.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df.tail()
```

### 1. TimeShifting

Documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shift.html

```python
df.shift(1) #  Loss of data at the end |Adds NaNs at the beginning of the dataframe.
```

```python
df.shift(-1) # Loss of data at the beginning| Adds NaNs to end of the dataframe
```

```python
# USing frequency to shift.
df.shift(periods=1, freq='M') # Last day of the month.
```

### 2. Rolling and Exapnding

```python
# Visualizing:
df['Close'].plot(figsize=(12,5))
```

```python
# Adding rolling mean
df['Close']
df.rolling(window = 7).mean() 

# In this case the 1st 7 entries will be NaNs.
# Reason: Not enough data to take average of window = 7 days.
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window = 30).mean() ['Close'].plot()
```

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window = 30).mean()
```

```python
df[['Close','Close: 30 Day Mean']].plot(figsize=(12,5))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))

# last point is the overall average of the whole dataset.
```

```python
df.expanding().mean().plot(figsize=(12,5))
```
