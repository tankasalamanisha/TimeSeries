## Data Generating Process(DGP):

Synthetic data generation process is important to understand as the model needs to learn from all possible inputs and outcomes. There are many techniques of generating this synthetic data.

1. White and red noise.

A. White Noise: It has a sequence of random numbers with zero as mean and a constant standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(200)
values = np.random.randn(200)*100
```

```python
def plot_time_series(x,y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid()
    plt.show()
```

```python
plot_time_series(x=time, y=values, title="White Noise")
```

B. Red noise:
It has zero mean and constant variance but is serially correlated in time. This serial correlation or redness is parameterized by a correlation coefficient r, such that:

$${x}_{j+1} = {r}.{x}_{j} + {(1 - r^{2})}^{1/2}.{w}$$

```python
r = 0.4
time = np.arange(200)
white_noise = np.random.randn(200)*100
values = np.zeros(200)
for i, v in enumerate(white_noise):
    if i == 0:
        values[i] = v
    else:
        values[i] = r*values[i-1] + np.sqrt(1- np.power(r,2))*v

plot_time_series(time, values, "Red Noise")
```
