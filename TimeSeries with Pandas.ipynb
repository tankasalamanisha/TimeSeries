```python
# percentage increase from the beginning.
idx = pd.IndexSlice
df['%increase'] = 0
previous = df.loc[df.index[0],idx['Volume']]
next_value=0
for timestamp in df.index[1:]:
    next_value = df.loc[timestamp,idx['Volume']]
    df.loc[timestamp,idx['%increase']] = 100 * (previous - next_value)/previous
    previous= df.loc[timestamp,idx['Volume']]
```

```python
df.head()
```

```python
# percentage decrease from the beginning.
idx = pd.IndexSlice
df['%decrease'] = 0
previous = df.loc[df.index[0],idx['Volume']]
next_value=0
for timestamp in df.index[1:]:
    next_value = df.loc[timestamp,idx['Volume']]
    df.loc[timestamp,idx['%decrease']] = 100 * (next_value-previous)/next_value
    previous= df.loc[timestamp,idx['Volume']]
```

## Statsmodels: Understanding and application

Time Series has the following components associated withit:
1. Trends: Gradual increase or decrease in the values. Increase represents the "upward trend" and decrease represents the "downward trend".

2. Seasonality: Repeating trends as in in a year/month/week etc, the values tend to repeat themselves in a constant pattern.

3. Cyclical : Trends with no set repetition. Its not known to occur in a known repetitive cycle.

### Hodrick Prescott filter:
 It seperates a time-series into a trend and cyclical component.
 
 y_t = T_t + C_t
 
 These components are determined by minimizing the quadratic loss function.

 ${min}_{\tau _{t}} \sum \limits _{t=1} ^{\tau} ={𝑐}_{t} ^{2} {Tt} + {\lambda}\sum \limits _{t= 1} ^{T} [({\tau}_{t}−{\tau}_{t−1})−({\tau}_{t−1}−{\tau}_{t−2})]^{2} $

 ${\lambda}$ -> Smoothing parameter


 There are some default values that can be used for lambda. 

 Quarterly data -> 1600 <br>
 Annual data -> 6.25<br>
 Monthly data -> 129,600<br>
